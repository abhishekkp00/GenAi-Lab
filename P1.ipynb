```python
import gensim.downloader as api #used to load pretrained model
```

```python
import numpy as np
```

```python
from numpy.linalg import norm  #used to calculate the vector magnitude
```

```python
# Load GloVe pre-trained word vectors (100-dimensional)
# This model is trained on Wikipedia + Gigaword corpus
model = api.load("glove-wiki-gigaword-100")

print("Vocabulary size:", len(model.key_to_index))
```

```python
similar_words = model.most_similar("king", topn=5)

print("\nTop 5 words similar to 'king':")
for word, score in similar_words:
    print(f"{word} : {score}")
```

```python
# Perform vector arithmetic
result = model.most_similar(positive=["king", "woman"],
                            negative=["man"],
                            topn=1)

print("\nResult of king - man + woman:")
print(result)
```

```python
# Get vectors for individual words
king_vec = model["king"]      # Vector representation of 'king'
man_vec = model["man"]        # Vector representation of 'man'
woman_vec = model["woman"]    # Vector representation of 'woman'
```

```python
# Perform arithmetic manually
new_vector = king_vec - man_vec + woman_vec

# Find most similar word to the new vector
manual_result = model.similar_by_vector(new_vector, topn=5)

print("\nManual vector arithmetic result (Top 5):")
for word, score in manual_result:
    print(f"{word} : {score}")
```
